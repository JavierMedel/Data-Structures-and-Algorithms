```
A = [0, 0, -5, 30212]
B = [-10, 40, -3, 9]
V = -8
```

```
def sumOfTwo(A, B, V):

  complement = 0
  dir_comp = []

  for i in range(len(A)):
    complement = V - A[i]
    dir_comp.append(complement)

  if V in dir_comp:
    return True
  else:
    return False

  print(dir_comp)
```

```
sumOfTwo(A, B, V)
```
