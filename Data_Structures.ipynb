## Create our own array

```
class MyArray():
  def __init__(self):
    self.length = 0
    self.data = []

  def get(self, index):
    return self.data[index]

  def push(self, item):
    self.data.insert(self.length, item)
    self.length += 1

  def pop(self):
    temp_item = self.length
    del self.data[self.length-1]
    self.length -= self.length

  def delete(self, index):
    temp_item = self.data[index]
    self.shiftItems(index)
    return temp_item
    
  def shiftItems(self, index): # O(n)
    for index in range(self.length-1):
      self.data[index] = self.data[index+1]
    
    del self.data[self.length-1]
    self.length -=1
```

```
myArray = MyArray()
```

```
myArray.push('hello')
myArray.push('you')
myArray.push('!')
```

```
myArray.get(2)
```

```
def reverse(string=''):
  if type(string) != str:
    print('its no a string')
  elif len(string) < 2:
    print('Small array')
  else:    
    temp_string = []
    for i in range(len(string)):
      temp_string.append(string[(len(string) -1) - i])
    return ''.join(temp_string)
```

```
reverse('[4,5,6,7]')
```
