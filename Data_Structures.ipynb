## Arrays
* lookup O(1)
* append O(1)*
* insert O(n)
* delete O(n)

*can be O(n)

```
string = ['a','b','c','d']
```

```
# append // O(1)
string.append('e') # O(1)
string
```

```
# pop // O(1)
string.pop()
string
```

```
# insert  // O(n)
string.insert(3, 'x')
string
```

```
# remove // O(n)
string.remove('c')
string
```
